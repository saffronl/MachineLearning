```python
def run_linear(df):
     
    if df.shape[0] < 10: # 
        print("Not enough data to run linear regression model.")
    else:
        df_x = df[["RevenueUSD"]] #x is the Revenue
        df_y = pd.to_numeric(df["ClimateTotalEmissions"],errors="coerce") # y is the Total Emissions

        
        # train_test_split will split an array into random train and test subsets.
        #The training data is 80% of all the data, and it is used to train the model.
        #The test data is 20% of all the data, and is used to input into the model and get test results.
        X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=100)
        
        
        # LinearRegression fits a linear model with coefficients
        # Need to make intercept = false, since a company with 0 revenue should have 0 emissions!
        # Need to have postive = True, because a company cannot have negative emissions!
        lr = LinearRegression(fit_intercept = False, positive = True)
        lr.fit(X_train, y_train) # using the TRAINING model to create the linear regression equation
        # print the linear regression model variable:
        print("Coefficient: ",lr.coef_) #coef_ : array, shape (n_features, ) or (n_targets, n_features)
        
        # Now predict more values !
        y_lr_train_pred = lr.predict(X_train) # Predict using the linear model THIS IS TRAINING DATA
        y_lr_test_pred = lr.predict(X_test) # THIS IS TESTING DATA
        
        # Print results !!
        lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
        lr_train_r2 = r2_score(y_train, y_lr_train_pred)
        #print("Training Mean2 Error: ", y_lr_train_pred)
        #print("Training r2:          ", lr_train_r2)
        lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
        lr_test_r2 = r2_score(y_test, y_lr_test_pred)
        #print("Test Mean2 Error: ", lr_test_mse)
        #print("Test r2:          ", lr_test_r2)
```
